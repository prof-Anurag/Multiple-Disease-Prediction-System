# ***Import Dependencies***



```python
import pandas as pd
import numpy as np
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data Collection and **Processing**

# **0**  --> ***Represent***  ***Malignant***  
# ***1*** --> ***Represent*** ***Benign***

```python
# Loading the dataset from sklearn.datasets
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
print(breast_cancer_dataset)
```

```python
# Loading the data to a data frame.
data_frame=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
# Adding the 'target' column to the data frame.
data_frame['label']=breast_cancer_dataset.target
```

```python
# Number of Rows and Columns in the dataset.
data_frame.shape
```

```python
data_frame.tail()
# 0 --> Malignant
# 1 --> Benign
```

```python
# Getting some information about the dataset
data_frame.info()
```

```python
# Checking for Missing Values.
data_frame.isnull().sum()
```

```python
# Statistical Measures about the data.
data_frame.describe()
```

```python
# checking the Distribution of Target Variable.
data_frame['label'].value_counts()
```

```python
data_frame.groupby('label').mean()
```

```python
# Confusion Matrix
```

# ***Separate the Features and Target.***


```python
x=data_frame.drop(columns='label',axis=1)
y=data_frame['label']
```

```python
x.shape
```

```python
x
```

```python
y
```

# Split the Data into Training and Testing ***Data***

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

# ***MODEL*** ***TRAINING***

# LOGISTIC *REGRESSION*

```python
model=LogisticRegression()
```

```python
#Training the dataset using the Logistic Regression.
model.fit(x_train,y_train)
```

# ***Model Evaluation***

```python
# ACCURACY SCORE ON TRAINING DATA
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(y_train,x_train_prediction)
```

```python
print('Accuracy on training data :',training_data_accuracy)
```

```python
# ACCURACY SCORE ON TEST DATA.
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(y_test,x_test_prediction)
```

```python
print('Accuracy on test data :',test_data_accuracy)
```

## Generate a confusion matrix

```python
# Confusion matrix.
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
cm=confusion_matrix(y_test,x_test_prediction)
print(cm)
```

```python
# Plot heatmap of confusion matrix.
plt.figure(figsize=(5,4))
sns.heatmap(cm,annot=True,fmt='d',cmap='Blues',xticklabels=['Malignant','Benign'],
            yticklabels=['Malignant','Benign'])
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix Heatmap - Breast Cancer Dataset')
plt.show()
```
